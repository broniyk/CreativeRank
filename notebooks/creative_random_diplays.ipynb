### Display Random Images

```python
def display_random_images(image_folder="../images", num_images=8, rows=4, cols=2):
    """
    Randomly samples images from a folder and displays them in a grid.

    Parameters:
        image_folder (str): Path to the folder containing images.
        num_images (int): Number of images to sample and display.
        rows (int): Number of rows in the display grid.
        cols (int): Number of columns in the display grid.
    """
    # Supported image extensions
    exts = (".jpg", ".jpeg", ".png", ".gif", ".webp", ".tiff")
    # List all image files in the folder
    all_images = [
        os.path.join(image_folder, f)
        for f in os.listdir(image_folder)
        if f.lower().endswith(exts)
    ]
    if len(all_images) == 0:
        print(f"No images found in {image_folder}")
        return

    # Randomly sample images
    sample_images = random.sample(all_images, min(num_images, len(all_images)))

    # Plot images
    fig, axes = plt.subplots(rows, cols, figsize=(cols * 4, rows * 4))
    axes = axes.flatten()
    for ax, img_path in zip(axes, sample_images):
        print(img_path)
        try:
            img = Image.open(img_path)
            ax.imshow(img)
            ax.set_title(os.path.basename(img_path), fontsize=8)
            ax.axis('off')
        except Exception as e:
            ax.axis('off')
            ax.set_title("Error loading image", fontsize=8)
    # Hide any unused subplots
    for ax in axes[len(sample_images):]:
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
display_random_images("../images", num_images=8, rows=4, cols=2)
```
